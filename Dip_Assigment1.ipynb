```python
from PIL import Image
im = Image.open("xyz.jpg").convert('L')
im.save('greyscale.png')
im
```

```python
height , weight = im.size
im.size
```

```python
import numpy as np
img_array = np.array(im)
img_array.shape
```

```python
#flip
i = 0
j = 0
dt = img_array.dtype
if(weight > height):
    result = np.zeros([weight,weight],dt)
else:
    result = np.zeros([height,height],dt)
x = height-1
for i in range(weight):
    for j in range(height):
        result[j][x] = img_array[i][j]
    x-=1
result
img = Image. fromarray(result, 'L')
img.save('flip.png')
img
```

```python
#negative

negative = np.zeros([weight,height],dt)
negative.shape
for i in range(weight):
    for j in range(height):
        negative[i][j] = 255 - img_array[i][j]
img = Image. fromarray(negative, 'L')
img.save('negative.png')
img
```

```python
##bit planes
def Binary(pixel,depth):
        binary = []
        while(pixel > 0):
            binary.append(pixel%2)
            pixel = pixel // 2
        for i in range(8-len(binary)):
            binary.append(0)
        return binary[depth]
    
for i in range(weight):
        bitImage = ""
        for j in range(height):
            bitImage +=  str(Binary(img_array[i][j],3))
        print(bitImage)
        
```

```python
##contrast value
import math
sumz = 0
contrast1 = 0
brighness = 0
no_pixel = weight * height

for i in range(weight):
    for j in range(height):
        sumz += img_array[i][j]
brighness = sumz / no_pixel

for i in range(weight):
    for j in range(height):
        contrast1 +=  pow(img_array[i][j]-brighness , 2)
contrast = math.sqrt((1/no_pixel)*contrast1)
contrast
```

```python
#average intensity value
avg_intensity = np.zeros([weight,height],dt)
avg = brighness
for i in range(weight):
    for j in range(height):
        if(img_array[i][j] > avg):
            avg_intensity[i][j] = 255
        else:
            avg_intensity[i][j] = 0           
img = Image.fromarray(avg_intensity, 'L')
img.save('avg_intensity.png')
img
```

```python
#power law transformation

power_law = np.zeros([weight,height],dt)
gamz = input("Enter Value of gama ")
cz = input ("Enter Value of constant ")

gama = float(gamz)
c = int(cz)

for i in range(weight):
    for j in range(height):
        power_law[i][j] = (c*pow(img_array[i][j] , gama))
img = Image.fromarray(power_law, 'L')
img.save('power_law.png')
img
```

```python
# contrast stretching
contrast = np.zeros([weight,height],dt)
c = np.min(img_array)
d = np.max(img_array)
x = input("Enter lower bound to scratch ")
y = input("Enter max bound to scratch ")

a = int(x)
b = int(y)

for i in range(weight):
    for j in range(height):
        contrast[i][j] = round((img_array[i][j] - c) * ((b-a)/(d-c)) + a)
img = Image.fromarray(contrast, 'L')
img.save('contrast_stretching.png')
img
```

```python
##entropy
total_entropy = 0
minus = -1
(uni , frq) = np.unique(img_array , return_counts = True)
for i in range(len(frq)):
    total_entropy += ((frq[i] / (height* weight)) * (np.log2(frq[i] / (height* weight))))

print(minus*total_entropy)
```
